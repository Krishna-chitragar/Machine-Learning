```python
from keras.models import load_model
```

# Loading models

```python
model_1 = load_model('my_model.h5') # model with 5 epochs/batch size 100 from local 
model_2 = load_model('my_model_with_15_epochs.h5') # loading model with 15 epochs/batch size 32 with google colab
```

```python
Classes = ["Tomato_Bacterial_spot","Tomato_Early_blight","Tomato_Late_blight","Tomato_Leaf_Mold",
           "Tomato_Septoria_leaf_spot","Tomato_Spider_mites_Two_spotted_spider_mite","Tomato__Target_Spot",
           "Tomato__Tomato_YellowLeaf__Curl_Virus","Tomato__Tomato_mosaic_virus","Tomato_healthy"]
```

# Predictions


```python
import numpy as np
import matplotlib.pyplot as plt

# Pre-Processing test data same as train data.
img_width=256
img_height=256
#model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

from keras.preprocessing import image

def prepare(img_path):
    img = image.load_img(img_path, target_size=(256, 256))
    x = image.img_to_array(img)
    x = x/255
    return np.expand_dims(x, axis=0)
```

```python
result = model_1.predict_classes([prepare("late_blight.JPG")])
disease=image.load_img("late_blight.JPG")
plt.imshow(disease)
print (Classes[int(result)])
```
